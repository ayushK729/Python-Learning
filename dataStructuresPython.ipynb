# Data Structures in Python

- List : ordered, **mutable** collection of elements of various types,can contain duplicates
- Tuple : ordered, **immutable** list of elements of various types,can contain duplicates
- Dictionary : stores elements in key-pair format, the key must be unique and immutable, value can be duplicate and can be modified
- Set : unordered collection of unique elements, no duplicates

```python
l1=[]
print(f"Empty list -> {l1} \n Type of data -> {type(l1)}")
```

```python
l1=[1,2,3,4,5]
print(f"Complete list ->  {l1}")
print(f"First element in list -> {l1[0]}")
print(f"Last element in list -> {l1[-1]}")
print(f"Second last element in list -> {l1[-2]}")
```

```python
# using loop to iterate a list
for i in l1:
    print(i)
```

### List Comprehension

```python
# we can use List comprehension to create list without manually declaring each value

# list of first 10 natural nos

l1=[i**2 for i in range(1,11,1)]
print(l1)
```

```python
# square of first 25 numbrs that are even and print them with their original nos
l2=[i**2 for i in range(1,25,1) if i%2==0]
print(l2)
```

```python
# nested list comprehension
l1=[1,2,3]
l2=['a','b','c']
l3=[(i,j) for i in l1 for j in l2]
print(l3)
```

```python
# performing list concatenation
l4=l1+l2
print(f"Concatenated list -> {l4}")
```

## Tuples

```python
t1=()
t2=tuple()
t=(1,2,3)

print(f"Data -> {t1} \t Data type -> {type(t1)}")
print(f"Data -> {t2} \t Data type -> {type(t2)}")
print(f"Data -> {t} \t Data type -> {type(t)}")
```

```python
# Tuples are immutable
t1.append(23)
```

```python
# however, we can convert a tuple to list, add value and then convert back to tuple
print(f"data type of {t} -> {type(t)}")
t=list(t)
print(f"data type of {t} -> {type(t)}")
t.append(90)
print(f"data type of {t} -> {type(t)}")
t=tuple(t)
print(f"data type of {t} -> {type(t)}")
print(t)
```

```python
t2=(91,92)
```

```python
# concatenation of tuples
print(f"Tuple 1 -> {t}\t Tuple 2 -> {t2}")
t3=t+t2
print(f"Tuple concatenated -> {t3}")
```

```python
# Mutiplicative duplication in tuples

print(t2)
print(t2*3)
```

```python
# immutable nature of a tuple
t4=(1,2,2,2,1,1,2,34,5,5,6)
print(t4)
```

```python
# Packaging and unpackaging of tuples
# creating a packaged tuple
pkg_tuple=1,2,3,"Ayush",True
print(pkg_tuple)
print(f"Data type of pkg_tuple -> {type(pkg_tuple)}")
```

```python
# unpacking the packed tuple
a,*b,c=pkg_tuple
print(a)
print(b,type(b))
print(c)
```
