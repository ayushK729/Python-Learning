```python
# default initialisation of data types in python
name=str()
age=int()
height=float()
status=bool()

print(f"Default value of name -> {name}")
print(f"Default value of age -> {age}")
print(f"Default value of height -> {height}")
print(f"Default value of status -> {status}")
```

# Control Flows in Python

### Conditional Statements

```python
n=int(input("Input a number of choice:"))

if n>10:
    print("Larger than 10")
```

```python
n=int(input("Input a number of choice:"))

if n>10:
    print("Larger than 10")
else:
    print("Smaller than 10")
```

```python
# if elif
n=int(input("Input a number of choice:"))

if n>10:
    print("Number is larger than 10")
elif n<10 and n>0:
    print("Number lies between 1 and 10")
else:
    print("Number is 0")
```

### Looping Statements

```python
# for loop
for i in range(10):
    print(i)
```

```python
for i in range(1,20,2):
    print(i)
```

```python
# Nested looping
for i in range(1,3,1):
    for j in range(1,4,1):
        print(i,j)
```

```python
i=10
while(i!=1):
    print(i)
    i-=1
```
